The nine audio feature columns were extracted as a NumPy matrix with shape (28356, 9).

This normalization ensures that all audio features contribute equally to subsequent mathematical analysis.

# Bonus 1 – Artist Audio Fingerprint

A fingerprint vector was created for each artist by computing the mean value of nine normalized audio features.

Cosine similarity was implemented manually using NumPy based on the mathematical definition without external machine learning libraries.

Three artist pairs were tested:

- Queen vs Martin Garrix = 0.9996  
- Don Omar vs Guns N' Roses = 1.0000  
- Drake vs Hardwell = 0.9996  

Surprisingly, all similarity values were extremely high, even for artists from very different genres.

This suggests that averaging audio features at the artist level reduces stylistic variation, causing artist vectors to become very similar in direction.

As a result, cosine similarity in this form does not strongly separate artists by genre.

This experiment demonstrates an important data science lesson: a mathematically correct model does not always produce highly discriminative results, and feature engineering choices strongly influence analytical outcomes.

```python
scaled_df = df.copy()

scaled_df[features] = X_scaled
```

```python
scaled_df.shape
```

```python
def artist_fingerprint(df, artist_name):
    
    artist_data = df[df["track_artist"] == artist_name]
    
    mean_vector = artist_data[features].mean().to_numpy()
    
    return mean_vector
```

```python
def cosine_similarity(A, B):
    
    dot_product = np.dot(A, B)
    
    norm_A = np.linalg.norm(A)
    
    norm_B = np.linalg.norm(B)
    
    return dot_product / (norm_A * norm_B)
```

```python
pairs = [
    ("Queen", "Martin Garrix"),
    ("Don Omar", "Guns N' Roses"),
    ("Drake", "Hardwell")
]

for artist1, artist2 in pairs:
    
    vec1 = artist_fingerprint(scaled_df, artist1)
    
    vec2 = artist_fingerprint(scaled_df, artist2)
    
    similarity = cosine_similarity(vec1, vec2)
    
    print(f"{artist1} vs {artist2}: {similarity:.4f}")
```
